# Criptoaritmética

```python
from busqueda import *
from ambientes import *
%matplotlib inline
```

Para analizar el proyecto y ver la solución decidimos empezar con un problema propuesto que fue: son+one=new. Esto con el fin de mostrar el funcionamiento del proyecto con un problema no tan complejo.

```python
prob = CriptoAritmetica(['son', 'one', 'new'])
```

```python
s = prob.estado_inicial
prob.pintar_estado(s)
```

```python
print("Acciones aplicables:", prob.acciones_aplicables(s))
```

## Implementación de los algoritmos

Con los algoritmos de búsqueda vistos en clase (`best_first_search`, `breadth_first_search`, `depth_first_search`, `iterative_deepening_search`) procedemos a buscar la solución del problema planteado. Esto con el fin de observar la eficiencia de cada uno de estos y verificar la solución propuesta.

### Best First Search

```python
%%time
l = best_first_search(prob)
if l is not None:
    camino = solucion(l)
    print("La solución encontrada es: ", camino)
    print("El costo de la solución es: ", l.costo_camino)
else:
    print("No hay solución!")
```

```python
prob.pintar_camino(camino)
```

### Breadth First Search

```python
%%time
l2 = breadth_first_search(prob)
if l is not None:
    camino2 = solucion(l2)
    print("La solución encontrada es: ", camino2)
    print("El costo de la solución es: ", l2.costo_camino)
else:
    print("No hay solución!")
```

```python
prob.pintar_camino(camino2)
```

### Depth First Search

```python
%%time
l3 = depth_first_search(prob)
if l3 is not None:
    camino3 = solucion(l3)
    print("La solución encontrada es: ", camino3)
    print("El costo de la solución es: ", l3.costo_camino)
else:
    print("No hay solución!")
```

```python
prob.pintar_camino(camino3)
```

### Iterative Deeping Search

```python
%%time
l4 = iterative_deepening_search(prob, 100)
if l4 is not None:
    camino4 = solucion(l4)
    print("La solución encontrada es: ", camino4)
    print("El costo de la solución es: ", l4.costo_camino)
else:
    print("No hay solución!")
```

```python
prob.pintar_camino(camino4)
```

### Algoritmo con permutaciones

```python
%%time
camino5 = [(k, v) for k, v in prob.sol_algoritmo().items()]
if camino5 is not None:
    print("La solución encontrada es: ", camino5)
else:
    print("No hay solución!")
```

```python
prob.pintar_camino(camino5)
```

## Ejemplo Send + More = Money

Ahora realizamos el mismo procedimiento anteriormente hecho pero con el ejemplo más conocido en la criptoaritmética.

```python
prob = CriptoAritmetica(['send', 'more', 'money'])
s = prob.estado_inicial
prob.pintar_estado(s)
```

```python
print("Acciones aplicables:", prob.acciones_aplicables(s))
```

Al igual que el anterior ejemplo, utilizamos los algoritmos de búsqueda.

### Best First Search

```python
%%time
l = best_first_search(prob)
if l is not None:
    camino = solucion(l)
    print("La solución encontrada es: ", camino)
    print("El costo de la solución es: ", l.costo_camino)
else:
    print("No hay solución!")
```

```python
prob.pintar_camino(camino)
```

Dado que este problema tiene mayor complejidad y que los algoritmos de búsqueda no logran llegar a una solución en un tiempo no tan largo, se decidió establecer algunas condiciones iniciales para facilidad del usuario.

```python
s['d'] = 7
s['r'] = 8
prob.pintar_estado(s)
```

```python
print("Acciones aplicables:", prob.acciones_aplicables(s))
```

### Depth First Search

```python
%%time
l2 = depth_first_search(prob)
if l2 is not None:
    camino2 = solucion(l2)
    print("La solución encontrada es: ", camino2)
    print("El costo de la solución es: ", l2.costo_camino)
else:
    print("No hay solución!")
```

```python
camino2 += [('d', 7), ('r', 8)]
prob.pintar_camino(camino2)
```

### Iterative Deepening Search

```python
%%time
l3 = iterative_deepening_search(prob, 100)
if l3 is not None:
    camino3 = solucion(l3)
    print("La solución encontrada es: ", camino3)
    print("El costo de la solución es: ", l3.costo_camino)
else:
    print("No hay solución!")
```

```python
camino3 += [('d', 7), ('r', 8)]
prob.pintar_camino(camino3)
```

Como el algoritmo que requiere más tiempo es el de `breadth_first_search`, agregamos otra condición inicial para que el algoritmo logre llegar a la solución más pronto.

```python
s['m'] = 1
prob.pintar_estado(s)
```

### Breadth First Search

```python
%%time
l4 = breadth_first_search(prob)
if l4 is not None:
    camino4 = solucion(l4)
    print("La solución encontrada es: ", camino4)
    print("El costo de la solución es: ", l4.costo_camino)
else:
    print("No hay solución!")
```

```python
camino4 += [('d', 7), ('r', 8), ('m',1)]
prob.pintar_camino(camino4)
```

### Algoritmo con permutaciones

```python
%%time
camino5 = [(k, v) for k, v in prob.sol_algoritmo().items()]
if camino5 is not None:
    print("La solución encontrada es: ", camino5)
else:
    print("No hay solución!")
```

```python
prob.pintar_camino(camino5)
```
